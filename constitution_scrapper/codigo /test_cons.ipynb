```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from bs4 import NavigableString,Tag
import re

```

```python
# links paginas

url = "https://www.constituteproject.org/constitutions?lang=en&status=in_force&status=is_draft"
document = urlopen(url)

html = document.read()

soup = BeautifulSoup(html, "html.parser")

iterable = soup.find_all("a",href=True)

links = []

for x in iterable:
    if 'constitution' in x['href']:

        links.append(x['href'])

links = links[4:-1]
len(links)
```

```python
iterable
```

```python

# guarda el texto de las constituciones en data

host = "https://www.constituteproject.org"


for link in links:
    url = host + link

    document = urlopen(url)

    html = document.read()

    soup = BeautifulSoup(html, "html.parser")

    iterable = soup.find_all("p",attrs = {'class':'content'})

    constitution_name = link.split('/')[-1][:-8]

    filename ='../data/'+ constitution_name + '.txt'

    with open(filename,'w') as f:
        for x in iterable :
            f.write(x.getText())
            f.write("\n")
```
